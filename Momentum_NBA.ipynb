# Lets try a momentum start for just one player and then branch out to others. Our hypothesis is that players who are playing better than their average performance will continue to play better. Same idea as stocks. We will compute a moving average and try to model 1 step forward

```python
df1 = df[df["GID"] == 4379] # Try to do momentum strat for Kyrie as an example
pd.set_option('display.max_columns', 500)
df1.head(10)
```

```python
kyrie = df1[["First  Last", "Date", "Team pts", "Minutes", "FDP", "DKP", "FD Sal", "DK Sal"]]
kyrie.head()
```

```python
kyrie = kyrie.reindex(index=kyrie.index[::-1]) # reverse the order of the df to create appropriate prediction model
kyrie.head()
```

```python
kyrie = kyrie.drop(kyrie[kyrie.Minutes < 1].index)
```

# We will create a moving average (6 game) for FDP, DKP, FD Sal, DK Sal. We will also create a salary/points variable to see when kyrie is overpriced vs underpriced. 

```python
kyrie["Salary/Point FDP"] = kyrie["FD Sal"] / kyrie["FDP"] # Fan Duel salary per point
kyrie["Salary/Point DKP"] = kyrie["DK Sal"] / kyrie["DKP"] # draft kings salary per point

kyrie.head()
```

```python
# moving averages
kyrie['Salary/Point_FDP_SMA_6'] = kyrie["Salary/Point FDP"].rolling(window=6).mean()
kyrie['Salary/Point_DKP_SMA_6'] = kyrie["Salary/Point DKP"].rolling(window=6).mean()

kyrie.head(25)
```

```python
# RSI momnetum 
def computeRSI (data, time_window):
    diff = data.diff(1).dropna()        # diff in one field(one day)

    #this preservers dimensions off diff values
    up_chg = 0 * diff
    down_chg = 0 * diff
    
    # up change is equal to the positive difference, otherwise equal to zero
    up_chg[diff > 0] = diff[ diff>0 ]
    
    # down change is equal to negative deifference, otherwise equal to zero
    down_chg[diff < 0] = diff[ diff < 0 ]
    
    # check pandas documentation for ewm
    # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html
    # values are related to exponential decay
    # we set com=time_window-1 so we get decay alpha=1/time_window
    up_chg_avg   = up_chg.ewm(com=time_window-1 , min_periods=time_window).mean()
    down_chg_avg = down_chg.ewm(com=time_window-1 , min_periods=time_window).mean()
    
    rs = abs(up_chg_avg/down_chg_avg)
    rsi = 100 - 100/(1+rs)
    return rsi
```

```python
kyrie['6D_RSI'] = computeRSI(kyrie['Salary/Point FDP'], 5) 
kyrie['6D_RSI'] = computeRSI(kyrie['Salary/Point DKP'], 5) 
```

```python
kyrie.head(20)
```

# BUY WHEN RSI CROSSES OVER 50 for outperformance for next game

```python
# lets test the results!
kyrie["FDP Change"] = kyrie["FDP"].pct_change() * 100;
kyrie["DKP Change"] = kyrie["DKP"].pct_change() * 100;

kyrie.head(20)
```

 # Algo: If todays RSI is >= 50 buy the player for their next game. This is a momentum strat.

```python
final_result = kyrie[["First  Last", "Date", "FDP", "DKP", "6D_RSI", "DKP Change", "FDP Change"]]
final_result.head(20)
```

```python
final_result['Signal'] = np.where(final_result['6D_RSI'] >= 50, 1, 0) # is RSI >= to 50? If yes, buy for tomorrows game
final_result.head(20)
```
